## Activity 11: Retrieving Data Correctly from Databases

```python
# importing needed package
import sqlite3
```

### 1. Connect to petsDB and check whether the connection has been successful.

```python
connection = sqlite3.connect("petsdb")
```

```python
# function to test that the connection was successful. It executes the connection
# if it is able to execute it returns TRUE
# if the connection is closed, it returns FALSE
def opened(connection):
    try:
        connection.execute("SELECT * FROM persons LIMIT 1")
        return True
    except sqlite3.ProgrammingError as e:
        print("Connection closed {}".format(e))
        return False
```

```python
# did the connection work?
opened(connection)
```

We received a True so the connection was successful!

### 2. Find the different age groups in the persons database.

```python
# creating my cursor
cursor = connection.cursor()
```

```python
# this executes a query to display the count for each age group
for ppl, age in cursor.execute("SELECT count(*), age FROM persons GROUP BY age"):
    print("There are {} people aged {}".format(ppl, age))
```

### 3. Find the age group that has the maximum number of people.

### 4. Find the people who do not have a last name.

```python
# This loops through adn counts the number of last_names which are Null
for count in connection.execute("SELECT count(*) FROM persons WHERE last_name IS null"):
    print('There are {} people with no last name.'.format(count[0]))
```

### 5. Find out how many people have more than one pet.

```python
# this loops through and counts the number of people with more than one pet
for count in connection.execute("SELECT count(*) FROM (SELECT count(owner_id) FROM pets GROUP BY owner_id HAVING count(owner_id) >1)"):
    print("There are {} people with more than one pet.".format(count[0]))
```

### 6. Find out how many pets have received treatment.

```python
# this loops through and counts the number of pets who have received treatment
for count in connection.execute("SELECT count(*) FROM pets WHERE treatment_done=1"):
    print("There are {} pets who have received treatment.".format(count[0]))
```

### 7. Find out how many pets have received treatment and the type of pet is known.

```python
# this loops through and counts the number of pets who have a type and have received treatment
for count in connection.execute("SELECT count(*) FROM pets WHERE treatment_done=1 AND pet_type IS NOT null"):
    print("There are {} pets that have received treatment and the type is also known.".format(count[0]))
```

### 8. Find out how many pets are from the city called east port.

```python
# this loops through and counts the number of pets win the city east port
for count in connection.execute("SELECT count(*) FROM pets JOIN persons ON pets.owner_id = persons.id WHERE persons.city='east port'"):
    print("There are {} pets from East Port.".format(count[0]))
```

### 9. Find out how many pets are from the city called east port and who received a treatment.

```python
# this loops through and counts the number of pets who received treatment and are from east port
for count in connection.execute("SELECT count(*) FROM pets JOIN persons ON pets.owner_id = persons.id WHERE persons.city='east port' AND pets.treatment_done=1"):
    print("There are {} pets from East Port that received treatment.".format(count[0]))
```
